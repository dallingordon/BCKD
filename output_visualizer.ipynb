```python
import matplotlib.pyplot as plt
```

```python
import re

def extract_float(s):
    # This regex pattern looks for any sequence of digits (\d+), optionally followed by
    # a decimal point and more digits (\.\d+)? The entire pattern is wrapped in parentheses
    # to capture the match as a group.
    match = re.search(r'(\d+(\.\d+)?)', s)
    if match:
        return float(match.group(0))
    else:
        return None

def plot_metrics(file_path):
    # Initialize empty lists to store data
    epochs = []
    train_acc = []
    train_loss = []
    test_acc = []
    test_acc_top5 = []
    test_loss = []

    # Read the file line by line
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Parse data
    for line in lines:
        if line.startswith('epoch:'):
            epoch = int(line.split(':')[1].strip())
            epochs.append(epoch)
        elif line.startswith('train_acc:'):
            train_acc.append(float(line.split(':')[1].strip()))
        elif line.startswith('train_loss:'):
            train_loss.append(float(line.split(':')[1].strip()))
        elif line.startswith('test_acc:'):
            test_acc.append(float(line.split(':')[1].strip()))
        elif line.startswith('test_acc_top5:'):
            test_acc_top5.append(float(line.split(':')[1].strip()))
        elif line.startswith('test_loss:'):
            test_loss.append(float(line.split(':')[1].strip()))
        elif line.startswith('best_acc'):
            best_acc = extract_float(line.split()[1])
            
    if best_acc is not None:
        print(f"Best Accuracy: {best_acc}")
    else:
        print("no best printed, script did not finish!")

    plt.figure(figsize=(10, 8))
    plt.plot(epochs, train_loss, label='Train Loss')
    plt.plot(epochs, test_loss, label='Test Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Train and Test Loss over Epochs')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot train and test accuracies
    plt.figure(figsize=(10, 8))
    plt.plot(epochs, train_acc, label='Train Accuracy')
    plt.plot(epochs, test_acc, label='Test Accuracy')
    plt.plot(epochs, test_acc_top5, label='Test Accuracy Top5')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Train and Test Accuracies over Epochs')
    plt.legend()
    plt.grid(True)
    plt.show()
```
