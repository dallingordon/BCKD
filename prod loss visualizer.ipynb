```python
def create_2d_heat_map_with_contours(x_min, x_max, y_min, y_max, interval):
    """
    Generates a 2D heat map with contour lines, where the color intensity represents the product of its x and y values.
    
    Parameters:
    - x_min: Minimum value for the x-axis.
    - x_max: Maximum value for the x-axis.
    - y_min: Minimum value for the y-axis.
    - y_max: Maximum value for the y-axis.
    - interval: Interval between each point on both axes.
    """
    # Create a meshgrid for X and Y values
    x = np.arange(x_min, x_max, interval)
    y = np.arange(y_min, y_max, interval)
    X, Y = np.meshgrid(x, y)
    
    # Compute the product of X and Y values
    Z = X * Y
    
    # Plot
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Create the heat map
    heatmap = ax.contourf(X, Y, Z, cmap='RdYlGn', levels=100)
    plt.colorbar(heatmap)
    
    # Add contour lines to indicate levels of equal value
    contour = ax.contour(X, Y, Z, colors='black', linewidths=0.5)
    ax.clabel(contour, inline=True, fontsize=8)
    
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_title('2D Heat Map with Contour Lines of X * Y')
    plt.show()

# Example usage
create_2d_heat_map_with_contours(0.5, 1, 1, 2, 0.1)
```
